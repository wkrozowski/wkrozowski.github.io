```python
from matplotlib import pyplot as plt
import numpy as np
import sympy as sym
import math 
import random
```

```python
#define our function
x=sym.Symbol('x');
f=x**2

#integration range
a=0;
b=1;
```

```python
def montecarlo(f1,ax1,ax2,ay1,ay2,n):
    L0=[]; L1=[]; L_1=[]
    for i in range(n):
        ax=random.uniform(ax1,ax2);
        ay=random.uniform(ay1,ay2);
        if (ay<f1.subs(x,ax) and ay>0):
            L1.append((ax,ay));
        else:
            if (ay>f1.subs(x,ax) and ay<0):
                L_1.append((ax,ay));
            else: L0.append((ax,ay));
    #print(L1);
    #print(float(len(L1)-len(L_1))/n);
    #print(float((ax2-ax1)*(ay2-ay1)));
    I=((float(len(L1)-len(L_1))/n)*float((ax2-ax1)*(ay2-ay1)));
    Lx1, Ly1 = zip(*L1);
    plt.plot(Lx1, Ly1,marker='o', linestyle='None', markersize = 5.0);
    if L_1!=[]:
        Lx_1, Ly_1 = zip(*L_1);
        plt.plot(Lx_1, Ly_1,marker='o', linestyle='None', markersize = 5.0);
    Lx0, Ly0 = zip(*L0);
    plt.plot(Lx0, Ly0,marker='o', linestyle='None', markersize = 5.0);
    plt.show;
    print(I);
```

```python
montecarlo(f,0,3,-1,9,500);
```
