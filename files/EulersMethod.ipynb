```python
from matplotlib import pyplot as plt
import numpy as np
import sympy as sym
import math 
```

First, we define the value of our **parameters**. We can define here the number of iterations for our algorithm.

```python
#step size
h=0.1;
#Number of iterations
k=20
```

By using the syntax of Python's symbolic maths, we define the differential equation we want to study.
We must also define an **initial condition**, in the form of a pair of coordinates. We can define several sets of initial coordinates to study different solutions.

```python
#define our differential equation
x=sym.Symbol('x');
y=sym.Symbol('y');
dy=sym.Symbol('dy');
dy=y+x;

#set of initial conditions
x01=-10;
y01=-2;

#other set of initial conditions
x02=-8;
y02=-2;
```

We now define a function which takes initial conditions as an argument, and produces a list of coordinates obtained from the iterations of Euler's method.

```python
def processODE(x1,y1):
    L1=[x1];
    L2=[y1];
    for i in range(k):
        yn=L2[i]+h*dy.subs({x:L1[i], y:L2[i]})
        L1.append(float(L1[i]+h));
        L2.append(float(yn));
    return(L1,L2);
```

Now, we **plot the results** of the previous function, for the initial conditions defined above.

```python
LX1,LY1=processODE(x01,y01);
LX2,LY2=processODE(x02,y02);
plt.plot(LX1,LY1);
plt.plot(LX2,LY2);


plt.show()
```

To go a little further, we can now define a function that executes Euler's method with a **range of initial conditions**. For instance, f(0)=-10 to 10 (you can choose a different range and play around). This allows us to visualise the shape of many solutions at once.

```python
for i in range(-10,10,1):
    LX1,LY1=processODE(i,0);
    plt.plot(LX1,LY1);
```
